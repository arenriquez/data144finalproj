# Start Training Models

We'll start by training linear regression models to predict G3 grades (final grade) based on several features. We're going to construct models for each dataset (math and Portuguese), since there would be duplicate student entries if we were to combine the two datasets.

### Data Cleaning

```python
# Add wk_drinks and wkend_drinks columns to Portuguese dataset
port_copy1['wkday_drinks'] = port_copy1['Dalc'].apply(lambda x:'low' if x==1 else 'non-low')
port_copy1['wkend_drinks'] = port_copy1['Walc'].apply(lambda x:'low' if x==1 else 'non-low')
port_copy1.head()
```

### LinearRegression model for predicting grades on math student dataset

```python
# Initial features to train models on
grade_features = ['G1', 'G2', 'wkday_drinks', 'wkend_drinks', 'higher']

# Get training X and Y math datasets
math_grade_Xtrain = math_data[grade_features]
math_grade_Xtrain = pd.get_dummies(math_grade_Xtrain)
math_grade_Ytrain = math_data['G3']

# Split training datasets into training and validation data
mg_X_train, mg_X_val, mg_Y_train, mg_Y_val = train_test_split(math_grade_Xtrain, math_grade_Ytrain)
```

```python
# Test model on validation data
math_grade_model = LinearRegression()
math_grade_model.fit(mg_X_train, mg_Y_train)
math_grade_score = r2_score(np.array(mg_Y_val), math_grade_model.predict(mg_X_val))
math_grade_score
```

```python
print("The validation accuracy of our LinearRegression model to predict G3 grades for our math student ")
print("dataset was approximately: " + str(round(math_grade_score*100, 2)) + "%.")
```

```python
lm = LinearRegression()
scores = cross_val_score(lm, math_grade_Xtrain, math_grade_Ytrain, scoring='r2', cv=5)     
```

```python
print(np.mean(scores))
```

```python
print("The validation accuracy of our LinearRegression model to predict G3 grades for our math student ")
print("dataset was approximately: " + str(np.mean(scores)*100) + "%.")
```

### Decision tree for predicting 'higher' in math student dataset

```python
# Initial features to train models on
higher_features = ['G1', 'G2', 'G3', 'wkday_drinks', 'wkend_drinks', 'reason', 'famsup']

# Get training X and Y math datasets
math_higher_Xtrain = math_data[higher_features]
math_higher_Xtrain = pd.get_dummies(math_higher_Xtrain)
math_higher_Ytrain = math_data['higher']
```

```python
# Initialize model
math_higher_model = DecisionTreeClassifier()
```

```python
# 1000 iterations to see range of our model's accuracy
mh_scores = []
for i in np.arange(1000):
    mh_X_train, mh_X_val, mh_Y_train, mh_Y_val = train_test_split(math_higher_Xtrain, math_higher_Ytrain)
    math_higher_model.fit(mh_X_train, mh_Y_train)
    math_higher_score = accuracy_score(math_higher_model.predict(mh_X_val), np.array(mh_Y_val))
    mh_scores.append(math_higher_score)
display(min(mh_scores), max(mh_scores), np.mean(mh_scores))
```

```python
# Cross-validated classifier
mh_kfold = KFold()
mh_kfold.get_n_splits(math_higher_Xtrain)
mh_kfold_classifier = DecisionTreeClassifier()
for train_index, val_index in mh_kfold.split(math_higher_Xtrain):
    #print("TRAIN:", train_index, "VAL:", val_index)
    mh_Xtrain_kfold, mh_Xval_kfold = math_higher_Xtrain.loc[train_index], math_higher_Xtrain.loc[val_index]
    mh_Ytrain_kfold, mh_Yval_kfold = math_higher_Ytrain.loc[train_index], math_higher_Ytrain.loc[val_index]
    mh_kfold_classifier.fit(mh_Xtrain_kfold, mh_Ytrain_kfold)
    print("Training: " + str(accuracy_score(mh_kfold_classifier.predict(mh_Xtrain_kfold), mh_Ytrain_kfold)))
    print("Validation: " + str(accuracy_score(mh_kfold_classifier.predict(mh_Xval_kfold), mh_Yval_kfold)))

cross_val_score(math_higher_model, math_higher_Xtrain, math_higher_Ytrain, cv=5)
```

### Decision tree for predicting 'higher' with Portuguese student dataset

```python
# Get training X and Y port datasets
port_higher_Xtrain = port_copy1[higher_features]
port_higher_Xtrain = pd.get_dummies(port_higher_Xtrain)
port_higher_Ytrain = port_copy1['higher']
```

```python
# Initialize model
port_higher_model = DecisionTreeClassifier()
```

```python
# 1000 iterations to see range of our model's accuracy
ph_scores = []
for i in np.arange(1000):
    ph_X_train, ph_X_val, ph_Y_train, ph_Y_val = train_test_split(port_higher_Xtrain, port_higher_Ytrain)
    port_higher_model.fit(ph_X_train, ph_Y_train)
    port_higher_score = accuracy_score(port_higher_model.predict(ph_X_val), np.array(ph_Y_val))
    ph_scores.append(port_higher_score)
display(min(ph_scores), max(ph_scores), np.mean(ph_scores))
```

```python
# Cross-validated classifier
ph_kfold = KFold()
ph_kfold.get_n_splits(port_higher_Xtrain)
ph_kfold_classifier = DecisionTreeClassifier()
for train_index, val_index in ph_kfold.split(port_higher_Xtrain):
    #print("TRAIN:", train_index, "VAL:", val_index)
    ph_Xtrain_kfold, ph_Xval_kfold = port_higher_Xtrain.loc[train_index], port_higher_Xtrain.loc[val_index]
    ph_Ytrain_kfold, ph_Yval_kfold = port_higher_Ytrain.loc[train_index], port_higher_Ytrain.loc[val_index]
    ph_kfold_classifier.fit(ph_Xtrain_kfold, ph_Ytrain_kfold)
    print("Training: " + str(accuracy_score(ph_kfold_classifier.predict(ph_Xtrain_kfold), ph_Ytrain_kfold)))
    print("Validation: " + str(accuracy_score(ph_kfold_classifier.predict(ph_Xval_kfold), ph_Yval_kfold)))

cross_val_score(port_higher_model, port_higher_Xtrain, port_higher_Ytrain, cv=5)
```

```python
# Bagging
ph_scores_bagSets = {}
ph_kscores = []
ph_bag_tree_sets = []
for k in np.arange(15):
    bag_train_predicted = []
    bag_trees = [DecisionTreeClassifier() for i in np.arange(k+1)]
    for tree in bag_trees:
        indices = np.random.choice(np.arange(port_higher_Xtrain.shape[0]), port_higher_Xtrain.shape[0], replace = True)
        x_train_bag = port_higher_Xtrain.loc[indices]
        y_train_bag = port_higher_Ytrain.loc[indices]
        tree.fit(x_train_bag, y_train_bag)
        bag_train_predicted.append(tree.predict(port_higher_Xtrain))
    # print(bag_train_predicted)
    ph_bag_tree_sets.append(bag_trees)
    bag_train_predictions = stats.mode(bag_train_predicted)[0][0]
    ph_kscores.append(accuracy_score(bag_train_predictions, port_higher_Ytrain))
    ph_scores_bagSets[accuracy_score(bag_train_predictions, port_higher_Ytrain)] = k + 1
print(ph_kscores)
```
