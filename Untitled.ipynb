```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA

```python
# Display previews of datasets
math = pd.read_csv('student-mat.csv')
port = pd.read_csv('student-por.csv')
display(math.head(), port.head())
```

```python
# Display length of math and portuguese class datasets
print("There are " + str(math.shape[0]) + " entries in the math dataset.")
print("There are " + str(port.shape[0]) + " entries in the Portuguese dataset.")
```

```python
# All columns in dataset as a Numpy array
columns = np.array(math.columns)
columns
```

```python
# Combine both datasets for convenience; specify 'course' column for each dataset
# Is this useful? I don't know yet
math_copy = math.copy(); port_copy = port.copy()
math_copy['course'] = ['math' for i in np.arange(math.shape[0])]
port_copy['course'] = ['portuguese' for i in np.arange(port.shape[0])]
combined = pd.concat([math_copy, port_copy])
combined
```

```python
# Group by school, sex, and age individually (Portuguese dataset, since there are more entries)
display(port.groupby("school").count()[["sex"]], 
        port.groupby("sex").count()[["school"]], 
        port.groupby("age").count()[["school"]])
```

```python
# Histogram of ages in Portuguese dataset
plt.hist(port['age'], bins=8)
plt.title("Ages in Portuguese Course")
plt.xlabel("Age"); plt.ylabel("Count")
```

```python
# Group by failures
port.groupby("failures").count()[['school']]
```

```python
# Group by Dalc and Walc individually
display(port.groupby("Dalc").count()[["school"]], port.groupby("Walc").count()[["school"]])
```

```python
# Group by Dalc and failures together
port.groupby(["Dalc", "failures"]).count()[['school']]
```

```python
# Group by goout and Dalc
port.groupby(['goout', 'Dalc']).count()[['school']]
```
